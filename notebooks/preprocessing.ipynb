```python
import pandas as pd
import numpy as np

def clean_sector_closures(
    filepath: str,
    sector_name: str,
    min_months: int = 6
) -> pd.DataFrame:
    """
    Cleans a Statistics Canada business dynamics dataset for a single sector.

    Steps:
    - Load raw CSV
    - Drop junk first row (row 0)
    - Standardize column names
    - Remove footnotes using numeric-density rule
    - Keep only 'Closing businesses'
    - Convert monthly data to numeric
    - Reshape wide  long
    - Aggregate monthly  yearly
    - Create 1-year lag feature

    Returns:
    DataFrame with columns:
    industry | year | closures | closures_lag1
    """

    # -----------------------------
    # 1. Load raw data
    # -----------------------------
    df = pd.read_csv(filepath, skiprows=11, encoding="utf-8-sig")

    # Drop first junk row (contains 'number')
    df = df.replace("..", np.nan)
    df = df.dropna(axis=1, how='all')
    df = df.drop(0, axis=0).reset_index(drop=True)

    # -----------------------------
    # 2. Clean column names
    # -----------------------------
    df.columns = (
        df.columns.astype(str)
        .str.replace("\n", " ", regex=False)
        .str.replace(r"\s+", " ", regex=True)
        .str.strip()
    )

    # Identify ID column (measure name)
    id_col = df.columns[0]

    # Identify month columns
    month_cols = [c for c in df.columns if c != id_col]

    # -----------------------------
    # 3. Build numeric shadow copy
    #    (used ONLY to detect footnotes)
    # -----------------------------
    df_num = df.copy()

    for c in month_cols:
        df_num[c] = (
            df_num[c]
            .astype(str)
            .str.replace("\u00a0", "", regex=False)  # non-breaking space
            .str.replace(",", "", regex=False)       # thousands separator
            .str.replace(" ", "", regex=False)
            .replace({"..": None, "": None, "": None, "nan": None})
        )
        df_num[c] = pd.to_numeric(df_num[c], errors="coerce")

    # Keep rows with sufficient numeric coverage
    mask = df_num[month_cols].notna().sum(axis=1) >= min_months
    df_clean = df_num.loc[mask].copy()

    # -----------------------------
    # 4. Keep only "Closing businesses"
    # -----------------------------
    df_clean = df_clean[
        df_clean[id_col]
        .astype(str)
        .str.lower()
        .str.contains("closing")
    ].copy()

    # If nothing matched, return empty df safely
    if df_clean.empty:
        return pd.DataFrame(
            columns=["industry", "year", "closures", "closures_lag1"]
        )

    # -----------------------------
    # 5. Wide  Long (monthly)
    # -----------------------------
    df_long = df_clean.melt(
        id_vars=[id_col],
        value_vars=month_cols,
        var_name="month",
        value_name="closures"
    )

    # Ensure closures are numeric (critical)
    df_long["closures"] = pd.to_numeric(df_long["closures"], errors="coerce")

    # -----------------------------
    # 6. Extract year
    # -----------------------------
    df_long["year"] = (
        df_long["month"]
        .astype(str)
        .str.extract(r"(\d{4})")
        .astype(int)
    )

    # -----------------------------
    # 7. Aggregate monthly  yearly
    # -----------------------------
    yearly = (
        df_long
        .groupby("year", as_index=False)
        .agg(closures=("closures", "sum"))
        .sort_values("year")
    )

    # -----------------------------
    # 8. Create lag feature (no leakage)
    # -----------------------------
    yearly["closures_lag1"] = yearly["closures"].shift(1)

    # Drop first lag row 
    yearly = yearly.dropna(subset=["closures_lag1"]).copy()

    # -----------------------------
    # 9. Add industry label
    # -----------------------------
    yearly["industry"] = sector_name

    # -----------------------------
    # 10. Final column order
    # -----------------------------
    yearly = yearly[
        ["industry", "year", "closures", "closures_lag1"]
    ]

    return yearly
```

```python
import re

def normalize_industry(s):
    if pd.isna(s):
        return s
    s = str(s)
    s = re.sub(r"\s*\[[^\]]+\]\s*", "", s)  # remove [72], [31-33], etc
    s = re.sub(r"\s+", " ", s)
    return s.strip().lower()
```

```python
# Identify ID vs numeric columns
id_cols = ["naics_code", "industry", "industry_key","year"]
num_cols = [c for c in master_df.columns if c not in id_cols]

# Collapse quarterly duplicates  yearly
master_df = (
    master_df
    .groupby(id_cols, as_index=False)[num_cols]
    .mean(numeric_only=True)
)
```

```python
master_df.head()
```

```python
# make sure naics_code is stringy and clean
master_df["naics_code_str"] = master_df["naics_code"].astype(str).str.extract(r"(\d+)")[0]

# sector = first 2 digits
master_df["sector_code"] = master_df["naics_code_str"].str[:2]
```

```python
# collapse 31/32/33 -> 31 (Manufacturing)
master_df["sector_code"] = master_df["sector_code"].replace({"32": "31", "33": "31"})

# collapse 44/45 -> 44 (Retail Trade)
master_df["sector_code"] = master_df["sector_code"].replace({"45": "44"})

# collapse 48/49 -> 48 (Transportation and Warehousing)
master_df["sector_code"] = master_df["sector_code"].replace({"49": "48"})
```

```python
sector_name_map = (
    master_df.groupby("sector_code")["industry"]
    .agg(lambda s: s.mode().iloc[0] if not s.mode().empty else s.iloc[0])
    .to_dict()
)

master_df["sector"] = master_df["sector_code"].map(sector_name_map)
```

```python
# choose your numeric feature columns (everything except ids/text)
id_cols = ["sector_code", "sector", "year"]
target_col = "insolvencies"

feature_cols = [
    c for c in master_df.columns
    if c not in ["naics_code", "naics_code_str", "industry", "industry_key", "quarter_num", target_col]
    and master_df[c].dtype != "O"
]

final_df = (
    master_df
    .groupby(id_cols, as_index=False)
    .agg(
        insolvencies=(target_col, "sum"),
        **{c: (c, "mean") for c in feature_cols}
    )
)
```
